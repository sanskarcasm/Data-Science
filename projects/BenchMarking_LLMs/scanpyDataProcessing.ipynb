```python
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

def process_dataset(file_path, tissue_type, species):
    print(f"Processing {tissue_type} dataset: {file_path}")
    
    # Construct the full path to the file
    full_path = os.path.join("data", file_path)
    print(f"Attempting to access file: {full_path}")
    
    # Check if the file exists
    if not os.path.exists(full_path):
        print(f"Error: File not found: {full_path}")
        return None
    
    # Read the data
    adata = sc.read_text(full_path, delimiter='\t', first_column_names=True)
    adata = adata.transpose()
    
    # Basic preprocessing
    sc.pp.filter_cells(adata, min_genes=200)
    sc.pp.filter_genes(adata, min_cells=3)
    
    # Normalize the data
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)
    
    # Identify highly variable genes
    sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
    
    # Scale the data
    sc.pp.scale(adata, max_value=10)
    
    # Perform PCA
    sc.tl.pca(adata, svd_solver='arpack')
    
    # Compute the neighborhood graph
    sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
    
    # Perform clustering
    sc.tl.leiden(adata)
    
    # Run UMAP
    sc.tl.umap(adata)
    
    # Find marker genes
    sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon')
    
    # Plot UMAP
    sc.pl.umap(adata, color='leiden', save=f"{tissue_type}_{species}_umap.png")
    
    # Get top 10 marker genes for each cluster
    top_genes = pd.DataFrame(adata.uns['rank_genes_groups']['names']).head(10)
    top_genes.to_csv(f"{tissue_type}_{species}_top_genes.csv")
    
    return adata

# List of datasets
datasets = [
    ("GSM4008619_Adult-Kidney2_dge.txt", "Adult_Kidney", "Human"),
    ("GSM4008620_Adult-Kidney3_dge.txt", "Adult_Kidney", "Human"),
    ("GSM4008623_Adult-Liver1-1_dge.txt", "Adult_Liver", "Human"),
    ("GSM4008624_Adult-Liver1-2_dge.txt", "Adult_Liver", "Human"),
    ("GSM4008628_Adult-Lung1_dge.txt", "Adult_Lung", "Human"),
    ("GSM4008629_Adult-Lung2_dge.txt", "Adult_Lung", "Human"),
    ("GSM4008678_Fetal-Brain3_dge.txt", "Fetal_Brain", "Human"),
    ("GSM4008679_Fetal-Brain4_dge.txt", "Fetal_Brain", "Human"),
    ("GSM4008693_Fetal-Kidney3_dge.txt", "Fetal_Kidney", "Human"),
    ("GSM4008694_Fetal-Kidney4_dge.txt", "Fetal_Kidney", "Human"),
    ("GSM4409674_mouse_AdultArenalGland_1_dge.txt", "Adult_Kidney", "Mouse"),
    ("GSM4409675_mouse_AdultArenalGland_2_dge.txt", "Adult_Kidney", "Mouse"),
    ("GSM4409683_mouse_AdultStomach_2_dge.txt.gz", "Adult_Lung", "Mouse")
]

# In the main part of your script:
for file_name, tissue_type, species in datasets:
    adata = process_dataset(file_name, tissue_type, species)
    if adata is not None:
        adata.write(f"{tissue_type}_{species}_{file_name.split('_')[0]}.h5ad")
    else:
        print(f"Skipping {file_name} due to error.")

print("Processing complete!")
```

```python
import scanpy as sc
import pandas as pd
import os

# Function to process a single dataset
def process_dataset(exp_matrix_path, cell_info_path, output_prefix):
    print(f"Processing dataset: {exp_matrix_path} with cell info: {cell_info_path}")
    
    # Construct full paths
    full_exp_matrix_path = os.path.join('data', exp_matrix_path)
    full_cell_info_path = os.path.join('data', cell_info_path)
    
    # Check if files exist
    if not os.path.exists(full_exp_matrix_path):
        print(f"Error: Expression matrix file not found: {full_exp_matrix_path}")
        return None
    if not os.path.exists(full_cell_info_path):
        print(f"Error: Cell information file not found: {full_cell_info_path}")
        return None
    
    # Load the expression matrix
    adata = sc.read_text(full_exp_matrix_path, delimiter='\t', first_column_names=True)
    adata = adata.transpose()
    
    # Load the cell information
    cell_info = pd.read_csv(full_cell_info_path, delimiter='\t')
    
    # Ensure the cell information matches the cells in the expression matrix
    adata.obs = cell_info.set_index(adata.obs.index)
    
    # Basic preprocessing
    sc.pp.filter_cells(adata, min_genes=200)
    sc.pp.filter_genes(adata, min_cells=3)
    
    # Normalize the data
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)
    
    # Identify highly variable genes
    sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
    
    # Scale the data
    sc.pp.scale(adata, max_value=10)
    
    # Perform PCA
    sc.tl.pca(adata, svd_solver='arpack')
    
    # Compute the neighborhood graph
    sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
    
    # Perform clustering
    sc.tl.leiden(adata)
    
    # Run UMAP
    sc.tl.umap(adata)
    
    # Find marker genes
    sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon')
    
    # Plot UMAP
    sc.pl.umap(adata, color='leiden', save=f"_{output_prefix}_umap.png")
    
    # Get top 10 marker genes for each cluster
    top_genes = pd.DataFrame(adata.uns['rank_genes_groups']['names']).head(10)
    top_genes.to_csv(f"{output_prefix}_top_genes.csv")
    
    # Save the processed data
    adata.write(f"{output_prefix}_processed.h5ad")
    
    return adata

# List of datasets
datasets = [
    ("GSE235675_normalized_exp_matrix_bench1_scRNAseq_.txt", "GSE235675_cell_information_bench1_scRNAseq_.txt", "bench1"),
    ("GSE235675_normalized_exp_matrix_bench2_scRNAseq_.txt", "GSE235675_cell_information_bench2_scRNAseq_.txt", "bench2")
]

# Process each dataset
for exp_matrix_path, cell_info_path, output_prefix in datasets:
    adata = process_dataset(exp_matrix_path, cell_info_path, output_prefix)
    if adata is not None:
        print(f"Successfully processed {output_prefix}")
    else:
        print(f"Failed to process {output_prefix}")

print("Processing complete!")
```

```python
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image

def process_dataset(file_path, tissue_type, species):
    print(f"Processing {tissue_type} dataset: {file_path}")
    
    # Construct the full path to the file
    full_path = os.path.join("data", file_path)
    print(f"Attempting to access file: {full_path}")
    
    # Check if the file exists
    if not os.path.exists(full_path):
        print(f"Error: File not found: {full_path}")
        return None
    
    # Read the data
    adata = sc.read_text(full_path, delimiter='\t', first_column_names=True)
    adata = adata.transpose()
    
    # Basic preprocessing
    sc.pp.filter_cells(adata, min_genes=200)
    sc.pp.filter_genes(adata, min_cells=3)
    
    # Normalize the data
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)
    
    # Identify highly variable genes
    sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
    
    # Scale the data
    sc.pp.scale(adata, max_value=10)
    
    # Perform PCA
    sc.tl.pca(adata, svd_solver='arpack')
    
    # Compute the neighborhood graph
    sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
    
    # Perform clustering
    sc.tl.leiden(adata)
    
    # Run UMAP
    sc.tl.umap(adata)
    
    # Find marker genes
    sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon')
    
    # Plot UMAP with dataset name in the title
    umap_title = f"UMAP of {tissue_type} ({species}) - {file_path}"
    umap_save_name = f"{tissue_type}_{species}_umap.png"
    sc.pl.umap(adata, color='leiden', title=umap_title, save=f"_{umap_save_name}")
    
    return umap_save_name

# List of datasets
datasets = [
    ("GSM4008619_Adult-Kidney2_dge.txt", "Adult_Kidney", "Human"),
    ("GSM4008623_Adult-Liver1-1_dge.txt", "Adult_Liver", "Human"),
    ("GSM4008628_Adult-Lung1_dge.txt", "Adult_Lung", "Human"),
    ("GSM4008678_Fetal-Brain3_dge.txt", "Fetal_Brain", "Human"),
    ("GSM4008693_Fetal-Kidney3_dge.txt", "Fetal_Kidney", "Human"),
    ("GSM4409674_mouse_AdultArenalGland_1_dge.txt", "Adult_Kidney", "Mouse"),
]

# In the main part of your script:
umap_images = []
for file_name, tissue_type, species in datasets:
    umap_file = process_dataset(file_name, tissue_type, species)
    if umap_file is not None:
        umap_images.append(f"figures/{umap_file}")
    else:
        print(f"Skipping {file_name} due to error.")

print("UMAP plots generated. Creating collage...")

# Function to create a collage of UMAP images
def create_collage(images, collage_name, rows, cols, thumb_width=300, thumb_height=300):
    collage_width = cols * thumb_width
    collage_height = rows * thumb_height
    collage_image = Image.new('RGB', (collage_width, collage_height), (255, 255, 255))

    for i, image_file in enumerate(images):
        img = Image.open(image_file)
        img = img.resize((thumb_width, thumb_height), Image.ANTIALIAS)
        x_offset = (i % cols) * thumb_width
        y_offset = (i // cols) * thumb_height
        collage_image.paste(img, (x_offset, y_offset))

    collage_image.save(collage_name)
    print(f"Collage saved as {collage_name}")

# Create a collage with 3 columns and 4 rows (change accordingly if needed)
create_collage(umap_images, "umap_collage.png", rows=4, cols=3)

print("Collage creation complete!")
```

```python
import scanpy as sc
import pandas as pd
import os

def extract_top_markers(adata, n_genes=10):
    if 'rank_genes_groups' not in adata.uns:
        sc.tl.rank_genes_groups(adata, groupby='leiden', method='wilcoxon')
    
    top_genes = pd.DataFrame(
        {group: adata.uns['rank_genes_groups']['names'][group][:n_genes]
         for group in adata.obs['leiden'].cat.categories}
    )
    
    return top_genes

# List of datasets to process
datasets = [
    'bench2_processed.h5ad',
    'bench1_processed.h5ad',
    'Adult_Kidney_Mouse_GSM4409675.h5ad',
    'Adult_Kidney_Mouse_GSM4409674.h5ad',  # Corrected file name
    'Fetal_Kidney_Human_GSM4008693.h5ad',
    'Fetal_Brain_Human_GSM4008678.h5ad',
    'Adult_Lung_Human_GSM4008628.h5ad',
    'Adult_Liver_Human_GSM4008623.h5ad',
    'Adult_Kidney_Human_GSM4008619.h5ad'
]

# Process each dataset
for dataset in datasets:
    print(f"Processing {dataset}")
    try:
        adata = sc.read(dataset)
        top_genes = extract_top_markers(adata)
        output_file = dataset.replace('.h5ad', '_top_genes.csv')
        top_genes.to_csv(output_file)
        print(f"Saved top genes to {output_file}")
    except FileNotFoundError:
        print(f"Error: File not found: {dataset}")
    except Exception as e:
        print(f"Error processing {dataset}: {str(e)}")

print("Processing complete!")import os
import asyncio
import nest_asyncio
import pandas as pd
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

nest_asyncio.apply()

class LLMInterface:
    def __init__(self, model_name, token=None):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForCausalLM.from_pretrained(model_name)
        self.model_name = model_name

    async def query(self, prompt):
        inputs = self.tokenizer.encode(prompt, return_tensors="pt")
        outputs = self.model.generate(inputs, max_length=150)
        response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        return response

async def benchmark_models(models, prompts):
    results = {}
    for model_name in models:
        llm = LLMInterface(model_name)
        model_results = {}
        for prompt in prompts:
            response = await llm.query(prompt)
            model_results[prompt] = response
        results[model_name] = model_results
    return results

def load_dataset(file_path):
    return pd.read_csv(file_path)

def generate_prompts(dataset):
    prompts = []
    for _, row in dataset.iterrows():
        cluster = row['cluster']
        genes = row['marker_genes'].split(', ')
        prompt = f"Based on these marker genes for cluster {cluster}: {', '.join(genes)}, what cell type is this most likely to be?"
        prompts.append(prompt)
    return prompts

async def process_datasets(file_paths, models):
    for file_path in file_paths:
        print(f"Processing {file_path}...")
        dataset = load_dataset(file_path)
        prompts = generate_prompts(dataset)
        results = await benchmark_models(models, prompts)
        print(f"Results for {file_path}:")
        for model, model_results in results.items():
            print(f"Results for {model}:")
            for prompt, response in model_results.items():
                print(f"Prompt: {prompt}")
                print(f"Response: {response}\n")

async def main():
    models = [
        "meta-llama/Llama-2-7b-chat-hf",  # LLaMA model
        "DBRX/DBRX-instruct-beta-7b",  # DBRX model (replace with actual model name)
        "EleutherAI/gpt-neo-1.3B"  # Open-source model
    ]
    
    # List of dataset files in the current directory
    dataset_files = [
        "Adult_Kidney_Human_top_genes.csv",
        "Adult_Kidney_Mouse_top_genes.csv",
        "Adult_Liver_Human_top_genes.csv",
        "Adult_Lung_Human_top_genes.csv",
        "bench1_top_genes.csv",
        "bench2_top_genes.csv",
        "Fetal_Brain_Human_top_genes.csv",
        "Fetal_Kidney_Human_top_genes.csv"
    ]
    
    await process_datasets(dataset_files, models)

# Run the main function
asyncio.run(main())

```
